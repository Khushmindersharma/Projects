```python
# Number of matches won by the difending or chasing.
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("IPL 2022.csv")

data['won_by'] = data['won_by'].map({'Wickets' :'Chasing',
                                     'Runs'    :'Defending'})

won_by = data["won_by"].value_counts()
label = won_by.index
counts = won_by.values
colors = ['red','lightgreen']

fig = go.Figure(data =[go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = "Number of matches won by Defending or Chasing")
fig.update_traces(hoverinfo = "label+percent", textinfo = "value", 
                 textfont_size = 30,
                 marker = dict(colors = colors, 
                               line = dict(color = 'black',width = 3)))
fig.show()
```

```python
# Top scorer in IPL 2022. Bar graph is used for comparision.
# y = higher score, x = name of the players.

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("IPL 2022.csv")

figure = px.bar( data, 
                x = data["top_scorer"],
                y = data["highscore"],
               color = data["highscore"],
               title = "Top Scorers in IPL 2022")
figure.show()
```

```python
# Seeing what most teams perfer to batting or fielding after winning the toss.

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("IPL 2022.csv")

toss = data["toss_decision"].value_counts()
label = toss.index
counts = toss.values
colors = [ "skyblue","yellow"]

figure = go.Figure(data=[go.Pie(labels=label,
                               values=counts)])
figure.update_layout(title_text="Toss Decision")
figure.update_traces(hoverinfo ="label+percent",
                     textinfo="value",
                     textfont_size=30,
                     marker=dict(colors=colors,
                                 line=dict(color="black",
                                           width=3)))
figure.show()
```

```python
figure = px.bar(data,
                x = data["top_scorer"],
                title="Top scorers in IPL 2022")
figure.show()
```

```python
figure = go.Figure()
figure.add_trace(go.Bar(
    x=data["venue"],
    y=data["first_ings_wkts"],
    name="First Innings Wickets",
    marker_color="gold"))

figure.add_trace(go.Bar(
    x=data["venue"],
    y=data["second_ings_wkts"],
    name="Second Innings Wickets",
    marker_color="lightgreen"))

figure.update_layout(barmode="group",xaxis_tickangle=45)
figure.show()
```
